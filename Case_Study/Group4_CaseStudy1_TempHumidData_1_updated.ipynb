```python
def removal_of_outliers(df, column, threshold, up):
    if up == 1:
        removed_outliers = df[df[column] <= threshold]
    else:
        removed_outliers = df[df[column] >= threshold]

    sns.boxplot(removed_outliers[column])
    plt.title('Box Plot of ' + column)
    return removed_outliers
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train.values.reshape(-1, 1), y_train)
```

```python
y_pred_test = linear_model.predict(X_test.values.reshape(-1, 1))
y_pred_train = linear_model.predict(X_train.values.reshape(-1, 1))
```

```python
mean_squared_error(y_test, y_pred_test)
```

```python
mean_absolute_error(y_test, y_pred_test)
```

```python
r2_score(y_test, y_pred_test)
```

#**Interpretation of MSE, MAE and R^ in Linear Regression:**

* **MSE** : ***0.872*** suggests that, on average, the squared error (the square of the difference between actual and predicted y values) is 0.872 though lower values indicate better model performance, whether this is "good" depends on the scale of y

* **MAE** : ***0.747***, meaning that, on average, the absolute difference between the predicted and actual y values is 0.747 units. It has the same units as y and less sensitive to the outliers which is good

* **R^2 Score** :The R^2 score is ***0.928***, indicating that 92.8% of the variance in y_test is explained by our model then again, R^2 value that's a bit close to 1.0 means it's a good fit since it captures the linear relationship of x and y

The combination of a high R^2 score **(0.928)** and relatively low error values **(MSE: 0.872, MAE: 0.747)** indicates that the model performs well in capturing the linear relationship between the features (X) and the target (y).
